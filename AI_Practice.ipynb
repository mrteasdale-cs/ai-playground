```python
import os
import tarfile
from six.moves import urllib

import pandas as pd

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

'''
# This block of code checks if the directory specified by `housing_path` exists.
# If it does not exist, the directory is created using `os.makedirs()`.
# Then, the script downloads a compressed file (housing.tgz) from the URL specified by `housing_url`
# and saves it to the newly created directory. The downloaded file is opened as a tar archive,
# extracted into the directory, and then closed to release resources.
'''
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
  if not os.path.isdir(housing_path):
    os.makedirs(housing_path)
    tgz_path = os.path. join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()


def load_housing_data(housing_path=HOUSING_PATH):
  csv_path = os.path.join(housing_path, "housing.csv")
  return pd.read_csv(csv_path)

#call function
fetch_housing_data()
```

```python
import numpy as np

# Function to split data into training and testing sets based on a specified ratio
def split_data(data, ratio):
    # Set a random seed for reproducibility (ensures the same shuffle order every time the code is run)
    np.random.seed(14)
    # Shuffle the indices of the data randomly
    shuffled = np.random.permutation(len(data))

    set_size = int(len(data) * ratio)
    # Select indices for the test set (first 'set_size' indices from shuffled data) using list indexing
    test_indices = shuffled[:set_size]
    train_indices = shuffled[set_size:]

    return data.iloc[train_indices], data.iloc[test_indices]

# Split the dataset 'housing' into training (80%) and testing (20%) sets
train_set, test_set = split_data(housing, 0.2)

print(len(train_set), "train +", len(test_set), "test")
train_set.to_csv("train.csv")
test_set.to_csv("test.csv")
```

## Labels in a Supervised classification problem

1.   Dependent on Features: Labels are the outcomes that the model predicts based on input features.
2.   Categorical Nature: Categorical so they belong to a finite set of discrete classes (e.g., "spam" or "not spam").
3.   Explicitly Provided: In supervised learning, labels are explicitly provided during training. They guide the model in learning the relationship features and predictions.
4.   Used for Model Evaluation: Labels are critical for evaluating model performance through accuracy, precision, recall, and F1-score.
5. Influence on Generalization: High-quality labels help reduce noise in the dataset, enabling better generalisation on new data.
6. Potential Bias: Biased labels can propagate inequities into the model's predictions, making fairness and bias evaluation important considerations.

## Labels in Unsupervised learning:
1. No Labels: Data lacks predefined labels.
2. Pattern Discovery: Identifies hidden structures or clusters.
3. Exploratory: Focuses on finding relationships without guidance.
4. No Ground Truth: Evaluation is less straightforward.

## Categorical or Continuous?
Labels can be categorical (for classification tasks, e.g., "spam" or "not spam") or continuous (for regression tasks, e.g., predicting house prices)

## Linear regression

Linear regression models the relationship between two variables by fitting a linear equation to observed data. This equation is obtained by minimizing the sum of squared errors, which measures the distance between the predicted values and the actual data points.

The equation of a straight line is defined by two parameters:
- Slope (m): Indicates the direction of the line (upward or downward trend).
- Intercept (c): Represents the height of the line on the y-axis, often referred to as bias in machine learning.

Fit the line using least squares

To find the combination of the slope and intercept parameters for a regression model, you need to compare the model predictions with the true measurements and compute a sum of the squares of the residuals. By minimising this sum with respect to the unknown parameters, you will be able to identify a good fit of the given data.

The method is referred to as the least squares (LS) and can be applied to multiple regression and non-linear models as well.

a linear model makes a prediction by simply computing a weighted sum of the input features, plus a constant called the bias term (also called the intercept term),




```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Set a random seed for reproducibility
np.random.seed(42)

# Generate random arrays for X and y
# Create a dataset with 100 samples and 2 features
X = np.random.rand(100, 2) * 10  # Random values between 0 and 10
true_coefficients = np.array([1, 2])  # True coefficients for the model
true_intercept = 3  # True intercept

# Generate y with some added noise
y = np.dot(X, true_coefficients) + true_intercept + np.random.normal(0, 1, 100)

# Fit the linear regression model
reg = LinearRegression().fit(X[0:], y)

# Print model parameters
print(f"\nEstimated Coefficients: {reg.coef_}")
print(f"Estimated Intercept: {reg.intercept_}")

predict = reg.predict(X)

# Scatter plot of the first feature vs y
plt.scatter(X[:, 0], y, color='blue', alpha=0.5, label='Data Points')

# Generate predictions while keeping the second feature constant (e.g., its mean)
X_fixed = np.copy(X)
X_fixed[:, 1] = np.mean(X[:, 1])  # Fix the second feature to its mean value
predicted_y = reg.predict(X_fixed)

# Sort X[:, 0] and corresponding predictions for a smooth line
sorted_indices = np.argsort(X[:, 0])
plt.plot(X[sorted_indices, 0], predicted_y[sorted_indices],
         color="red", linewidth=3, label='Regression Line')
```

## Decision Trees
A decision tree has a flowchart-like structure:
1. Each internal node represents a ‘test’ on an attribute (for example, age < 30?).
2. Each branch represents the outcome of the test, and each leaf node represents a class label (decision taken after computing all attributes).
3. The paths from root to leaf represent classification rules.

Like SVMs, Decision Trees are versatile Machine Learning algorithms that can perform both classification and regression tasks, and even multioutput tasks.

```python
# import libraries
import pydotplus
import pandas as pd
import numpy as np
from sklearn import tree
from io import StringIO
import pydot
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

# load the dataset
iris = load_iris()

X = iris.data[:, 2:] # petal length and width - only 2 columns
y = iris.target

#pd_iris = pd.DataFrame(X, columns=['petal length', 'petal width'])
#pd_iris['target'] = y
```

```python
# Split the data set into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# fit the decision tree
tree_clf = DecisionTreeClassifier(criterion='entropy', max_depth=3)
tree_clf.fit(X_train, y_train)
```

```python
from sklearn.tree import export_graphviz
from IPython.display import Image

dot_out_file = export_graphviz(
    tree_clf,
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    filled = True,
    rounded = True
)

graph = pydotplus.graph_from_dot_data(dot_out_file)
Image(graph.create_png())
```

```python
# Computer some performance metrics
y_pred = tree_clf.predict(X_test)

# Print classification report
print(classification_report(y_test, y_pred, target_names=['setosa', 'versicolor', 'virginica']))

c_matrix = confusion_matrix(y_test, y_pred)

# Print confusion matrix
display = ConfusionMatrixDisplay(confusion_matrix=c_matrix, display_labels=['setosa', 'versicolor', 'virginica'])
display.plot()
plt.show()
```

# Week 4 K-Means and PCA


K-means is a popular and efficient partitional clustering algorithm that divides a dataset
D = \'${x^1,x^2,…,x^n}$' into *k* user-specified clusters, each with a centroid.

It groups similar data points into the same cluster and separates dissimilar ones into different clusters.

Despite its weaknesses, K-means remains widely used due to its simplicity.

The time complexity of K Means is O(tkn), where n is the number of data points, k is the number of clusters, and t is the number of iterations. Since both *k* and *t* are small, k-means is considered a linear algorithm.

```python
#Example
from sklearn.cluster import KMeans
import numpy as np

X = np.array([[1,2],[1,4],[1,0],
             [10,2],[10,4],[10,0]])

kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
print(kmeans.predict([[0,0],[12,3]]))
```

```python
# Import libraries
import pandas as pd
import numpy as np
import copy
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate a list of 2D data points on the plane
df = pd.DataFrame({
    'x': [12, 20, 28, 18, 29, 73, 24, 45, 45, 52, 51, 52, 55, 53, 77, 61, 64, 69, 72],
    'y': [39, 36, 30, 52, 54, 86, 55, 59, 63, 70, 66, 63, 58, 23, 77, 8, 19, 7, 24]
})
print(df.head())

# Plot the dataset
fig = plt.figure(figsize=(5, 5))
plt.scatter(df['x'], df['y'], color='k')
plt.xlim(0, 80)
plt.ylim(0, 80)
plt.show()
```

```python
# specify seed
k = 3

# randomly select a see centroid for each cluster
np.random.seed(100)
centroids = {}
for i in range(k):
      centroids[i+1] = [np.random.randint(0, 80), np.random.randint(0, 80)]

print(centroids)

# plot the centroids
fig = plt.figure(figsize=(5,5))
plt.scatter(df['x'], df['y'], color='k')
colmap = {1:'r', 2:'g', 3:'b'}
for i in centroids.keys():
      plt.scatter(*centroids[i], color=colmap[i])

plt.show()
```

```python
# Implement the cluster membership assignment according to Euclidean distances
def assignment(df, centroids):
    # For each current centroid
    for i in centroids.keys():
        # Calculate the Euclidean distance from it to all the data points
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df['x'] - centroids[i][0]) ** 2
                + (df['y'] - centroids[i][1]) ** 2
            )
        )

    # Find the closest cluster centroid from each data point
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))

    # Set a different plot color for each cluster
    df['color'] = df['closest'].map(lambda x: colmap[x])

    return df

# Run the 1st iteration of cluster membership assignment
df = assignment(df, centroids)
print(df.head())

# Plot the dataset and cluster centroids using their indicative color
fig = plt.figure(figsize=(5, 5))
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(0, 80)
plt.ylim(0, 80)
plt.show()
```

```python
# Backup the previous record of cluster centroids
old_centroids = copy.deepcopy(centroids)

# Implement the update function
def update(centroids):
    # For each cluster centroid
    for i in centroids.keys():
        # Update its coordinates (x,y) to the mean of the data points belonging to that cluster
        centroids[i][0] = np.mean(df[df['closest'] == i]['x'])
        centroids[i][1] = np.mean(df[df['closest'] == i]['y'])
    return centroids

# Run the 1st iteration of centroids update using the previously assigned cluster membership
centroids = update(centroids)
print(centroids)

# Plot the dataset and the new centroids
fig = plt.figure(figsize=(5, 5))
ax = plt.axes()
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(0, 80)
plt.ylim(0, 80)

# And also plot an arrow from each old centroid to its current updated coordinates
for i in old_centroids.keys():
    old_x = old_centroids[i][0]
    old_y = old_centroids[i][1]
    dx = (centroids[i][0] - old_centroids[i][0]) * 0.75
    dy = (centroids[i][1] - old_centroids[i][1]) * 0.75
    ax.arrow(old_x, old_y, dx, dy, head_width=2, head_length=3, fc=colmap[i], ec=colmap[i])
plt.show()
```

```python
# Run the 2nd iteration of cluster membership assignment using the previously updated centroids
df = assignment(df, centroids)
print(df.head())

# Plot the results
fig = plt.figure(figsize=(5, 5))
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(0, 80)
plt.ylim(0, 80)
plt.show()
```

```python
# Continue until all assigned categories don't change any more
while True:
    # Backup the current cluster membership
    closest_centroids = df['closest'].copy(deep=True)

    # Update the centroids
    centroids = update(centroids)

    # Update the cluster membership
    df = assignment(df, centroids)

    # Test if the previous and the updated cluster membership differ
    if closest_centroids.equals(df['closest']):
        # If not, then it means convergence has been reached
        break

# Plot the results
fig = plt.figure(figsize=(5, 5))
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(0, 80)
plt.ylim(0, 80)
plt.show()
```

```python
# Obtain a fresh version of the same dataset as above
df = pd.DataFrame({
    'x': [12, 20, 28, 18, 29, 73, 24, 45, 45, 52, 51, 52, 55, 53, 77, 61, 64, 69, 72],
    'y': [39, 36, 30, 52, 54, 86, 55, 59, 63, 70, 66, 63, 58, 23, 77, 8, 19, 7, 24]
})

# Run K-Means to find 3 clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(df)

# Get the final centroids
centroids = kmeans.cluster_centers_
print(centroids)

# Predict the cluster membership of each data point based on its Euclidean distance from the centroids
labels = kmeans.predict(df)
print(labels)

# Plot the results
fig = plt.figure(figsize=(5, 5))
colmap = {1: 'r', 2: 'g', 3: 'b'}
plt.scatter(df['x'], df['y'], color='k')
for idx, centroid in enumerate(centroids):
    plt.scatter(*centroid, color=colmap[idx+1])
plt.xlim(0, 80)
plt.ylim(0, 80)
plt.show()
```

# Week 5 Perceptrons and SVM


A multilayer perceptron (MLP) is formulated from loose biological principles. It ‘learns’ from data.

The learning process in MLP can be generally organised as:

*   Compute the outputs.
* Compare outputs with desired targets.
* Adjust the weights and repeat the process.

In addition, it can perform classification more accurately than other machine learning methods in applications, especially for high-dimensional data including 2D Scatter plots.

**Kernel** functions

In nonlinear SVM, you can use a kernel function as a means to perform the high-dimensional transformation. Common kernls include:
* linear
* polynomial of power p
* Gaussian (radial-basis function network)
* sigmoid.



```python
from sklearn import svm
import matplotlib.pyplot as plt
import numpy as np

X = [[0,0],[1,1]]
y = [0,1]

# Train the SVM model with a linear kernel
clf = svm.SVC(kernel='linear')
clf.fit(X, y)

# Create a grid of points to plot the decision boundary
xx, yy = np.meshgrid(np.linspace(-1, 2, 100), np.linspace(-1, 2, 100))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary and margins
plt.contour(xx, yy, Z, levels=[-1, 0, 1], linestyles=['--', '-', '--'], colors='k')

# Plot the training points
plt.scatter([x[0] for x in X], [x[1] for x in X], c=y, cmap=plt.cm.coolwarm, s=100)

# Highlight support vectors
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=150, facecolors='none', edgecolors='k', label="Support Vectors")

# Add labels and title
plt.title("SVM Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

* Raw image of 32 × 32 pixels as input

```python
# using the keras.utils.to_catagorical function
# convert the y_train to one hot encoding as per above image
y_train_onehot = to_categorical(y_train, 10)
y_test_onehot = to_categorical(y_test, 10)

print(f"The one hot for label '{y_train[230]}' is: {y_train_onehot[230]}")
```

```python
# convert images to values between 0 and 1 (normalise)
# to do this, we can divide by 255 due to each image having 255 pixel values
X_train = X_train.astype('float32') / 255
X_test = X_test.astype('float32') / 255

model = Sequential()

# add several layers to the CNN including the convelutional layers (32 = 32 depth/neurons)
# max pooling layers with 2x2 filters, and dropout layers with 0.25 / 25% probability
model.add(Input(shape=(32, 32, 3)))
model.add(Conv2D(32, (3,3), activation='relu', padding='same'))
model.add(Conv2D(32, (3,3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.125))
model.add(Conv2D(64, (3,3), activation='relu', padding='same'))
model.add(Conv2D(64, (3,3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(128, (3,3), activation='relu', padding='same'))
model.add(Conv2D(128, (3,3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
# at this point, our neurons are spatially arranged in a cube-like format rather
# than in just one row. To make this cube-like format of neurons into one row,
# we have to first flatten it.
model.add(Flatten())
# add a dense layet with 512 neurons
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(10, activation='softmax'))

model.summary()
```

```python
# compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# fit the model
trained = model.fit(X_train, y_train_onehot, batch_size=32, epochs=12, validation_split=0.2)
```

```python
# Install library “tqdm” from Anaconda Navigator
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import keras
from keras.layers import Dense, Dropout, Input
from keras.models import Model,Sequential
from keras.datasets import mnist
from tqdm import tqdm
from keras.layers import LeakyReLU
from keras.optimizers import Adam
```

```python
def load_data():
    # Load MNIST dataset and normalize the images
    (x_train, y_train), (x_test, y_test) = mnist.load_data()
    x_train = (x_train.astype(np.float32) - 127.5) / 127.5
    x_test = (x_test.astype(np.float32) - 127.5) / 127.5
    # Reshape images from (num_samples, 28, 28) to (num_samples, 784)
    x_train = x_train.reshape(-1, 784)
    x_test = x_test.reshape(-1, 784)
    return (x_train, y_train, x_test, y_test)

def adam_optimizer():
    return Adam(learning_rate=0.0002, beta_1=0.5)

def create_generator():
    generator = Sequential()
    generator.add(Dense(units=128, input_dim=100))
    generator.add(LeakyReLU(alpha=0.2))
    generator.add(Dense(units=128))
    generator.add(LeakyReLU(alpha=0.2))
    generator.add(Dense(units=784, activation='tanh'))
    generator.compile(loss='binary_crossentropy', optimizer=adam_optimizer())
    return generator

def create_discriminator():
    discriminator = Sequential()
    discriminator.add(Dense(units=128, input_dim=784))
    discriminator.add(LeakyReLU(alpha=0.2))
    discriminator.add(Dense(units=1, activation='sigmoid'))
    discriminator.compile(loss='binary_crossentropy', optimizer=adam_optimizer(), metrics=['accuracy'])
    return discriminator

def create_gan(discriminator, generator):
    # Freeze discriminator weights in the GAN model
    discriminator.trainable = False
    gan = Sequential()
    gan.add(generator)
    gan.add(discriminator)
    gan.compile(loss='binary_crossentropy', optimizer=adam_optimizer())
    return gan

def plot_generated_images(epoch, generator, examples=16, dim=(4, 4), figsize=(6, 6)):
    noise = np.random.normal(0, 1, [examples, 100])
    generated_images = generator.predict(noise)
    generated_images = generated_images.reshape(examples, 28, 28)
    plt.figure(figsize=figsize)
    for i in range(examples):
        plt.subplot(dim[0], dim[1], i + 1)
        plt.imshow(generated_images[i], interpolation='nearest', cmap='gray')
        plt.axis('off')
    plt.tight_layout()
    plt.savefig("gan_generated_image_epoch_%d.png" % epoch)
    plt.close()

def training(epochs=1, batch_size=128):
    # Load data
    (X_train, y_train, X_test, y_test) = load_data()
    batch_count = X_train.shape[0] // batch_size

    # Build GAN components
    generator = create_generator()
    discriminator = create_discriminator()
    gan = create_gan(discriminator, generator)

    for e in range(1, epochs + 1):
        print("Epoch %d" % e)
        for _ in tqdm(range(batch_count)):
            # Generate fake images
            noise = np.random.normal(0, 1, [batch_size, 100])
            generated_images = generator.predict(noise)
            # Select a random batch of real images
            idx = np.random.randint(0, X_train.shape[0], batch_size)
            image_batch = X_train[idx]
            # Combine real and fake images
            X = np.concatenate([image_batch, generated_images])
            y_dis = np.zeros(2 * batch_size)
            y_dis[:batch_size] = 1

            # Train discriminator
            discriminator.trainable = True
            discriminator.train_on_batch(X, y_dis)

            # Train GAN: try to trick the discriminator
            noise = np.random.normal(0, 1, [batch_size, 100])
            y_gen = np.ones(batch_size)
            discriminator.trainable = False
            gan.train_on_batch(noise, y_gen)

        # Periodically plot generated images
        if e == 1 or e % 20 == 0:
            plot_generated_images(e, generator)

training(epochs=200, batch_size=128)
```
